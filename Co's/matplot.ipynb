```python
#draw a line in a diagram from position (1,3) (2,10) (6,12) and (18,20)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x_point = np.array([1,2,6,18])
y_point = np.array([3,10,12,20])
plt.plot(x_point, y_point, marker="o", mec='g', mfc='g',color='r', linestyle='dotted')
plt.show()
```

```python
#two or more line on same plot
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([2,4,6,8,10])
x2 = np.array([1,2,3,4,5])
y1 = np.array([10,14,18,22,16])
y2 = np.array([10,20,30,40,50])

plt.plot(x1,y1,label='p')
plt.plot(x2,y2,label='q')
plt.xlabel("Speed")
plt.ylabel("Time")
plt.title("Two or more lines on same plot")
plt.legend()
plt.show()
```
