```python
#1
plt.suptitle("MARK OF DEPT")
x=np.array([1,2,3,5])
y=np.array([7,7,10,12])
plt.subplot(2,3,1)
plt.plot(x,y)
plt.grid(ls=':',color='olive',lw='0.9')
plt.title("bca")
#2
x=np.array([2,5,3,6])
y=np.array([5,7,4,1])
plt.subplot(2,3,2)
plt.plot(x,y)
plt.grid(ls=':',color='olive',lw='0.9')
plt.title("mca")
#3
x=np.array([0,1,3,5])
y=np.array([6,4,8,9])
plt.subplot(2,3,3)
plt.plot(x,y)
plt.grid(ls=':',color='olive',lw='0.9')
plt.title("btech")

#4
x=np.array([1,2,3,5])
y=np.array([7,7,10,12])
plt.subplot(2,3,4)
plt.plot(x,y)
plt.grid(ls=':',color='olive',lw='0.9')
plt.title("mtech")
#5
x=np.array([2,5,3,6])
y=np.array([5,7,4,1])
plt.subplot(2,3,5)
plt.plot(x,y)
plt.grid(ls=':',color='olive',lw='0.9')
plt.title("bsc")
#6
x=np.array([0,1,3,5])
y=np.array([6,4,8,9])
plt.subplot(2,3,6)
plt.plot(x,y)
plt.grid(ls=':',color='olive',lw='0.9')
plt.title("msc")
```

```python
#scatterFunction
#scatter1
x=np.array([0,1,2,3])
y=np.array([2,6,8,10])
c=['black','magenta','lime','olive']
plt.scatter(x,y,marker='o',color=c)
#scatter2
x=np.array([1,4,2,5])
y=np.array([3,5,7,9])
z=['hotpink','orange','cyan','teal']
plt.scatter(x,y,marker='o',color=z)
```

```python
#colormap
x=np.array([1,2,3,4])
y=np.array([5,6,7,8])
z=np.array([0,10,20,30])
plt.scatter(x,y,cmap='cividis',c=z)
plt.colorbar()
```

```python
x=np.array([1,2,3,4])
y=np.array([5,6,7,8])
sizes=np.array([20,30,80,100])
plt.scatter(x,y,s=sizes,alpha=0.5)
```

```python
#numpy_functions
from numpy import random
```

```python
x=np.random.randint(100,size=(100))
y=np.random.randint(100,size=(100))
colors=np.random.randint(100,size=(100))
sizes=10*random.randint(100,size=(100))
plt.scatter(x,y,c=colors,s=sizes,alpha=0.5,cmap='viridis')
plt.colorbar()
```

```python
x=np.array([1,2,3,5])
y=np.array([7,7,10,12])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Veritcal Bar Plot")
plt.bar(x,y,color='cyan',width=0.5)
```

```python
#horizontalBar
x=np.array([1,2,3,5])
y=np.array([7,9,10,12])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Horizontal Bar Plot")
plt.barh(x,y,color='hotpink',height=0.5)
```
